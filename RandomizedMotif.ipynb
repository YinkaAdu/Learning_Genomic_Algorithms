```python
def pattern_count(text, pattern):
    count = 0
    text_length = len(text)
    pattern_length = len(pattern)

    for i in range(text_length - pattern_length + 1):
        if text[i:i + pattern_length] == pattern:
            count += 1
    return count
```

```python
def frequent_words(text, k):
    frequent_patterns = set()  # Use a set to automatically remove duplicates
    counts = []
    text_length = len(text)

    for i in range(text_length - k + 1):
        pattern = text[i:i + k]
        counts.append(pattern_count(text, pattern))  # Use the pattern_count function

    max_count = max(counts)

    for i in range(text_length - k + 1):
        if counts[i] == max_count:
            frequent_patterns.add(text[i:i + k])

    return frequent_patterns
```

```python
def frequency_table(text, k):
    freq_map = {}
    n = len(text)
    for i in range(n - k + 1):
        pattern = text[i:i + k]
        if pattern not in freq_map:
            freq_map[pattern] = 1
        else:
            freq_map[pattern] += 1
    return freq_map
```

```python
def max_map(freq_map):

    if not freq_map:  # Handle the case of an empty dictionary
        return 0  # Or raise an exception, depending on the desired behavior
    return max(freq_map.values())
```

```python
def better_frequent_words(text, k):
   
    frequent_patterns = []
    freq_map = frequency_table(text, k)
    max_count = max_map(freq_map)

    for pattern, count in freq_map.items():
        if count == max_count:
            frequent_patterns.append(pattern)
    return frequent_patterns
```

```python
def clump_finder(text, k, L, t):
    
    frequent_k_mers = set()
    n = len(text)
    if n < L:
        return frequent_k_mers

    # 1. Compute the frequency table for the first window
    first_window = text[:L]
    freq_map = frequency_table(first_window, k)

    # 2. Check for clumps in the first window
    for kmer, count in freq_map.items():
        if count >= t:
            frequent_k_mers.add(kmer)

    # 3. Slide the window and update the frequency table
    for i in range(1, n - L + 1):
        # Remove the k-mer that is leaving the window
        old_kmer = text[i - 1:i - 1 + k]
        freq_map[old_kmer] -= 1
        if freq_map[old_kmer] == 0:
            del freq_map[old_kmer]  # Remove k-mer if its count reaches 0

        # Add the k-mer that is entering the window
        new_kmer = text[i + L - k:i + L]
        if new_kmer in freq_map:
            freq_map[new_kmer] += 1
        else:
            freq_map[new_kmer] = 1

        # Check for clumps in the current window
        if freq_map[new_kmer] >= t:
            frequent_k_mers.add(new_kmer)

    return frequent_k_mers
```

```python
def clump_finder(text, k, L, t):
    """
    Finds k-mers that form clumps in a DNA string.

    Args:
        text: The DNA string to search within.
        k: The length of the k-mers to find.
        L: The size of the clump window.
        t: The minimum number of times a k-mer must appear in a clump to be considered frequent.

    Returns:
        A set of k-mers that form clumps in the DNA string.
    """
    frequent_k_mers = set()
    text_length = len(text)

    for i in range(text_length - L + 1):
        window = text[i:i + L]
        kmer_counts = frequency_table(window, k)  # Use frequency_table to count k-mers in the window

        for kmer, count in kmer_counts.items():
            if count >= t:
                frequent_k_mers.add(kmer)

    return frequent_k_mers
```

```python
def dna_skewer_efficient(dna_string):
    """
    Computes the skew of a DNA string efficiently for large strings.

    Args:
        dna_string: The input DNA string.

    Returns:
        A space-separated string of skew values at each position.
    """
    skew_values = [0] * (len(dna_string) + 1)
    current_skew = 0
    for i, nucleotide in enumerate(dna_string):
        if nucleotide == 'G':
            current_skew += 1
        elif nucleotide == 'C':
            current_skew -= 1
        skew_values[i + 1] = current_skew
    return " ".join(map(str, skew_values))
```

```python
def dna_ori(dna_string):
    """
    Finds the minimum skew value and the positions of bases with that skew in a DNA string.

    Args:
        dna_string: The input DNA string.

    Returns:
        A dictionary where the key is the minimum skew value and the value is a list
        of the 0-based indices of the bases where that minimum skew occurs.
    """
    skew_values = [0]
    current_skew = 0
    for nucleotide in dna_string:
        if nucleotide == 'G':
            current_skew += 1
        elif nucleotide == 'C':
            current_skew -= 1
        skew_values.append(current_skew)

    min_skew = min(skew_values)
    min_skew_indices = [i for i, skew in enumerate(skew_values) if skew == min_skew]

    return {min_skew: min_skew_indices}
```

```python
def Hamming_dist(p, q):
    """
    Computes the Hamming distance between two approximately matched DNA strings.

    Args:
        p: The first DNA string.
        q: The second DNA string.

    Returns:
        The Hamming distance between the two strings (the number of positions
        at which the corresponding symbols are different).
        Returns -1 if the strings are not of equal length.
    """
    if len(p) != len(q):
        return -1  # Hamming distance is only defined for strings of equal length

    distance = 0
    for i in range(len(p)):
        if p[i] != q[i]:
            distance += 1
    return distance
```

```python
def reverse_complement(text):
    """
    Computes the reverse complement of a DNA string.

    Args:
        dna_string: The DNA string to be reversed and complemented.

    Returns:
        The reverse complement of the DNA string.
    """
    complement = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    reversed_string = text[::-1]  # Reverse the string
    reverse_complement_string = ''.join(complement[base] for base in reversed_string)
    return reverse_complement_string
```

```python
def neighbors(pattern, d):
    """
    Generates the neighborhood of a pattern with up to d mismatches.

    Args:
        pattern: The input string.
        d: The maximum number of mismatches.

    Returns:
        A set of all strings within Hamming distance d of pattern.
    """
    if d == 0:
        return {pattern}
    if len(pattern) == 1:
        return {'A', 'C', 'G', 'T'}
    neighborhood = set()
    suffix_neighbors = neighbors(pattern[1:], d)
    for suffix_neighbor in suffix_neighbors:
        for nucleotide in ['A', 'C', 'G', 'T']:
            if nucleotide == pattern[0]:
                neighborhood.add(nucleotide + suffix_neighbor)
            else:
                if d > 0:
                    neighborhood.add(nucleotide + suffix_neighbor)
    return neighborhood
```

```python
def hamming_distance(str1, str2):
    
    if len(str1) != len(str2):
        return -1  # Hamming distance is only defined for equal-length strings
    distance = 0
    for i in range(len(str1)):
        if str1[i] != str2[i]:
            distance += 1
    return distance

def approximate_pattern_count(text, pattern, d):
    """
    Counts the number of approximate occurrences of a pattern in a text,
    allowing up to d mismatches.

    Args:
        text: The text string to search within.
        pattern: The pattern string to search for.
        d: The maximum number of mismatches allowed.

    Returns:
        The number of times the pattern appears in the text with at most d mismatches.
    """
    count = 0
    for i in range(len(text) - len(pattern) + 1):
        pattern_prime = text[i:i + len(pattern)]  # Extract substring of text
        if hamming_distance(pattern, pattern_prime) <= d:
            count += 1
    return count
```

```python
def frequent_words_with_mismatches(text, k, d):
    
    freq_map = {}
    combined_text = text + reverse_complement(text)
    n = len(combined_text)
    for i in range(n - k + 1):
        pattern = combined_text[i:i + k]
        neighborhood = neighbors(pattern, d)
        for neighbor in neighborhood:
            if neighbor not in freq_map:
                freq_map[neighbor] = 1
            else:
                freq_map[neighbor] += 1
    m = max(freq_map.values()) if freq_map else 0  # Handle empty freq_map

    frequent_patterns = []
    for pattern, frequency in freq_map.items():
        if frequency == m:
            frequent_patterns.append(pattern)

    # Find the pattern(s) with the highest approximate_pattern_count
    max_approx_count = -1  # Initialize to a value that will be less than any count
    most_frequent_patterns = []

    for pattern in frequent_patterns:
        approx_count = approximate_pattern_count(combined_text, pattern, d)
        if approx_count > max_approx_count:
            max_approx_count = approx_count
            most_frequent_patterns = [pattern]  # Start a new list with the new most frequent pattern
        elif approx_count == max_approx_count:
            most_frequent_patterns.append(pattern)  # Add to the list if it's as frequent as the current most frequent

    return most_frequent_patterns # Return the list of most frequent patterns
```

```python
import itertools

def median_string(dna_input, pattern):
    """
    Calculates the sum of Hamming distances between a given k-mer pattern
    and a set of DNA strings.

    Args:
        dna_input: A space-separated string of DNA strings.
        pattern: The k-mer pattern string.

    Returns:
        The sum of Hamming distances between the pattern and the DNA strings.
    """

    def hamming_distance(str1, str2):
        """
        Calculates the Hamming distance between two strings.

        Args:
            str1: The first string.
            str2: The second string.

        Returns:
            The Hamming distance between the two strings.
        """
        if len(str1) != len(str2):
            raise ValueError("Strings must have equal length to calculate Hamming distance.")
        distance = 0
        for i in range(len(str1)):
            if str1[i] != str2[i]:
                distance += 1
        return distance

    def d(pattern, dna):
        """
        Calculates the total Hamming distance between a k-mer pattern and a list of DNA strings.
        d(Pattern, Dna)  The output is the sum of distances between Pattern and each string in Dna = {Dna1, ..., Dnat}.

        Args:
            pattern: The k-mer pattern.
            dna: A list of DNA strings.

        Returns:
            The total Hamming distance.
        """
        k = len(pattern)
        distance = 0
        for text in dna:
            min_dist = float('inf')
            for i in range(len(text) - k + 1):
                kmer = text[i:i + k]
                dist = hamming_distance(pattern, kmer)
                if dist < min_dist:
                    min_dist = dist
            distance += min_dist
        return distance

    dna_list = dna_input.split()
    return d(pattern, dna_list)
```

```python
import itertools

def median_string(dna_input, k):
    """
    Finds a k-mer pattern that minimizes the Hamming distance between the k-mers
    of the input DNA strings.  This is an implementation of the Median String problem.

    Args:
        dna_input: A space-separated string of DNA strings.
        k: The length of the k-mer to find.

    Returns:
        The k-mer pattern that minimizes the Hamming distance.
    """

    def hamming_distance(str1, str2):
        """
        Calculates the Hamming distance between two strings.

        Args:
            str1: The first string.
            str2: The second string.

        Returns:
            The Hamming distance between the two strings.
        """
        if len(str1) != len(str2):
            raise ValueError("Strings must have equal length to calculate Hamming distance.")
        distance = 0
        for i in range(len(str1)):
            if str1[i] != str2[i]:
                distance += 1
        return distance

    def d(pattern, dna):
        """
        Calculates the total Hamming distance between a k-mer pattern and a list of DNA strings.
        d(Pattern, Dna)

        Args:
            pattern: The k-mer pattern.
            dna: A list of DNA strings.

        Returns:
            The total Hamming distance.
        """
        k = len(pattern)
        distance = 0
        for text in dna:
            min_dist = float('inf')
            for i in range(len(text) - k + 1):
                kmer = text[i:i + k]
                dist = hamming_distance(pattern, kmer)
                if dist < min_dist:
                    min_dist = dist
            distance += min_dist
        return distance

    dna_list = dna_input.split()
    distance = float('inf')
    median = ""

    # Generate all possible k-mers: AA...AA to TT...TT
    nucleotides = ['A', 'C', 'G', 'T']
    all_kmers = [''.join(p) for p in itertools.product(nucleotides, repeat=k)]

    for pattern in all_kmers:
        current_distance = d(pattern, dna_list)
        if distance > current_distance:
            distance = current_distance
            median = pattern
    return median
```

```python
import itertools

def median_subroutine_string(dna, k):
    """
    Finds the median string from all possible k-mers.

    Args:
        dna: A space-separated string of DNA strings.
        k: The length of the k-mer.

    Returns:
        The median string.
    """

    def hamming_distance(str1, str2):
        """
        Calculates the Hamming distance between two strings.

        Args:
            str1: The first string.
            str2: The second string.

        Returns:
            The Hamming distance between the two strings.
        """
        if len(str1) != len(str2):
            raise ValueError("Strings must have equal length to calculate Hamming distance.")
        distance = 0
        for i in range(len(str1)):
            if str1[i] != str2[i]:
                distance += 1
        return distance

    def distance_between_pattern_and_strings(pattern, dna_list):
        """
        Calculates the total distance between a pattern and a list of DNA strings.
        This is equivalent to the d(Pattern, Dna) function.

        Args:
            pattern: The k-mer pattern.
            dna_list: A list of DNA strings.

        Returns:
            The sum of the minimum Hamming distances between the pattern
            and each string in dna_list.
        """
        k = len(pattern)
        total_distance = 0
        for text in dna_list:
            min_distance = float('inf')
            for i in range(len(text) - k + 1):
                kmer = text[i:i + k]
                dist = hamming_distance(pattern, kmer)
                if dist < min_distance:
                    min_distance = dist
            total_distance += min_distance
        return total_distance

    def all_strings(k):
        """
        Generates all possible strings of length k from the DNA alphabet (A, C, G, T).

        Args:
            k: The length of the strings to generate.

        Returns:
            A list of all possible k-mers.
        """
        nucleotides = ['A', 'C', 'G', 'T']
        return [''.join(p) for p in itertools.product(nucleotides, repeat=k)]

    dna_list = dna.split()
    distance = float('inf')
    median = ""
    patterns = all_strings(k)

    for pattern in patterns:
        current_distance = distance_between_pattern_and_strings(pattern, dna_list)
        if distance > current_distance:
            distance = current_distance
            median = pattern
    return median
```

```python
def profile_most_probable_kmer(text, k, profile):
    """
    Finds the profile-most probable k-mer in a given text.

    Args:
        text: A DNA string.
        k: The length of the k-mer.
        profile: A 4xK matrix representing the probability profile for each nucleotide (A, C, G, T) at each position.
                 profile[0] corresponds to A, profile[1] to C, profile[2] to G, and profile[3] to T.

    Returns:
        The profile-most probable k-mer in text.
    """

    def calculate_kmer_probability(kmer, profile):
        """
        Calculates the probability of a k-mer given a profile.

        Args:
            kmer: A DNA string of length k.
            profile: A 4xK matrix representing the probability profile.

        Returns:
            The probability of the k-mer.
        """
        probability = 1
        for i, nucleotide in enumerate(kmer):
            if nucleotide == 'A':
                probability *= profile[0][i]
            elif nucleotide == 'C':
                probability *= profile[1][i]
            elif nucleotide == 'G':
                probability *= profile[2][i]
            elif nucleotide == 'T':
                probability *= profile[3][i]
        return probability

    most_probable_kmer = ""
    max_probability = -1  # Initialize with a value that will be less than any valid probability

    for i in range(len(text) - k + 1):
        kmer = text[i:i + k]
        probability = calculate_kmer_probability(kmer, profile)
        if probability > max_probability:
            max_probability = probability
            most_probable_kmer = kmer

    return most_probable_kmer
```

```python
import numpy as np

def greedy_motif_search(dna_input, k, t):
    """
    Finds a set of k-mers that score optimally according to their consensus,
    using Laplace's Rule of Succession to construct the profile.

    Args:
        dna_input: A space-separated string of DNA strings.
        k: The length of the motif (k-mer).
        t: The number of DNA strings in Dna.

    Returns:
        A space-separated string representing the best motifs found.
    """

    def calculate_profile(motifs):
        """
        Calculates the profile matrix from a list of motifs, using Laplace's
        Rule of Succession.

        Args:
            motifs: A list of DNA strings, each of length k.

        Returns:
            A 4xK numpy array representing the probability profile.
            Rows are A, C, G, T.
        """
        k = len(motifs[0])
        profile = np.zeros((4, k))
        for i in range(k):
            counts = {'A': 0, 'C': 0, 'G': 0, 'T': 0}
            for motif in motifs:
                counts[motif[i]] += 1
            # Apply Laplace's Rule:
            profile[0][i] = (counts['A'] + 1) / (len(motifs) + 4)
            profile[1][i] = (counts['C'] + 1) / (len(motifs) + 4)
            profile[2][i] = (counts['G'] + 1) / (len(motifs) + 4)
            profile[3][i] = (counts['T'] + 1) / (len(motifs) + 4)
        return profile

    def profile_most_probable_kmer(text, k, profile):
        """
        Finds the profile-most probable k-mer in a given text.

        Args:
            text: A DNA string.
            k: The length of the k-mer.
            profile: A 4xK numpy array representing the probability profile.

        Returns:
            The profile-most probable k-mer in text.
        """
        most_probable_kmer = ""
        max_probability = -1

        for i in range(len(text) - k + 1):
            kmer = text[i:i + k]
            probability = 1
            for j, nucleotide in enumerate(kmer):
                if nucleotide == 'A':
                    probability *= profile[0][j]
                elif nucleotide == 'C':
                    probability *= profile[1][j]
                elif nucleotide == 'G':
                    probability *= profile[2][j]
                elif nucleotide == 'T':
                    probability *= profile[3][j]
            if probability > max_probability:
                max_probability = probability
                most_probable_kmer = kmer
        return most_probable_kmer

    def score(motifs):
        """
        Calculates the consensus score of a list of motifs.

        Args:
            motifs: A list of DNA strings, each of length k.

        Returns:
            The score (sum of the most frequent nucleotides at each position).
        """
        k = len(motifs[0])
        score = 0
        for i in range(k):
            counts = {'A': 0, 'C': 0, 'G': 0, 'T': 0}
            for motif in motifs:
                counts[motif[i]] += 1
            score += max(counts.values())
        return score

    dna_list = dna_input.split()
    best_motifs = [seq[:k] for seq in dna_list]
    best_score = score(best_motifs)

    for i in range(len(dna_list[0]) - k + 1):
        motif1 = dna_list[0][i:i + k]
        motifs = [motif1]
        for j in range(1, t):
            profile = calculate_profile(motifs[:j])
            motif_j = profile_most_probable_kmer(dna_list[j], k, profile)
            motifs.append(motif_j)
        current_score = score(motifs)
        if current_score < best_score:
            best_score = current_score
            best_motifs = motifs
    return ' '.join(best_motifs)
```

```python
import random
from collections import Counter

def random_kmer(sequence, k, profile):
    n = len(sequence)
    probabilities = []
    kmers = []
    for i in range(n - k + 1):
        kmer = sequence[i:i + k]
        prob = 1
        for j, nucleotide in enumerate(kmer):
            prob *= profile[nucleotide][j]
        kmers.append(kmer)
        probabilities.append(prob)

    total = sum(probabilities)
    probabilities = [p / total for p in probabilities]
    return random.choices(kmers, weights=probabilities, k=1)[0]

def build_profile(motifs, pseudocount=1):
    k = len(motifs[0])
    profile = {nuc: [pseudocount] * k for nuc in 'ACGT'}

    for motif in motifs:
        for idx, nucleotide in enumerate(motif):
            profile[nucleotide][idx] += 1

    for i in range(k):
        total = sum(profile[nuc][i] for nuc in 'ACGT')
        for nuc in 'ACGT':
            profile[nuc][i] /= total

    return profile

def score(motifs):
    k = len(motifs[0])
    score = 0
    for i in range(k):
        column = [motif[i] for motif in motifs]
        most_common = Counter(column).most_common(1)[0][1]
        score += len(motifs) - most_common
    return score

def gibbs_sampler_once(Dna, k, t, N):
    motifs = [random.choice([s[i:i+k] for i in range(len(s) - k + 1)]) for s in Dna]
    best_motifs = motifs[:]

    for j in range(N):
        i = random.randint(0, t - 1)
        motifs_except_i = motifs[:i] + motifs[i+1:]
        profile = build_profile(motifs_except_i, pseudocount=1)
        motifs[i] = random_kmer(Dna[i], k, profile)

        if score(motifs) < score(best_motifs):
            best_motifs = motifs[:]

    return best_motifs

def run_multiple_starts(dna_string, k, t, N, runs=20):
    Dna = dna_string.strip().split()
    best_overall = gibbs_sampler_once(Dna, k, t, N)
    best_score = score(best_overall)

    for _ in range(runs - 1):
        motifs = gibbs_sampler_once(Dna, k, t, N)
        current_score = score(motifs)
        if current_score < best_score:
            best_score = current_score
            best_overall = motifs

    return ' '.join(best_overall)
```

```python
import random

random.seed(0)  # <- force reproducible random choices

def random_kmer(text, k):
    start = random.randint(0, len(text) - k)
    return text[start:start + k]

def profile_matrix_with_pseudocounts(motifs):
    k = len(motifs[0])
    # Initialize with pseudocounts (Laplace rule: start with 1 everywhere)
    profile = {'A': [1] * k, 'C': [1] * k, 'G': [1] * k, 'T': [1] * k}

    for motif in motifs:
        for idx, nucleotide in enumerate(motif):
            profile[nucleotide][idx] += 1  # Add actual counts

    total_counts = len(motifs) + 4  # +4 because we added 1 pseudocount per nucleotide
    for nucleotide in profile:
        for i in range(k):
            profile[nucleotide][i] /= total_counts  # Normalize by total

    return profile

def profile_most_probable_kmer(text, k, profile):
    max_prob = -1
    best_kmer = text[0:k]

    for i in range(len(text) - k + 1):
        kmer = text[i:i+k]
        prob = 1
        for j, nucleotide in enumerate(kmer):
            prob *= profile[nucleotide][j]
        if prob > max_prob:
            max_prob = prob
            best_kmer = kmer

    return best_kmer

def score_motifs(motifs):
    score = 0
    k = len(motifs[0])

    for i in range(k):
        column = [motif[i] for motif in motifs]
        most_common = max(set(column), key=column.count)
        score += sum(1 for nucleotide in column if nucleotide != most_common)

    return score

def randomized_motif_search_one_iteration_with_initial_motifs(dna_strings, k, t, initial_motifs_str):
    dna = dna_strings.strip().split()
    initial_motifs = initial_motifs_str.strip().split()

    if len(initial_motifs) != t:
        raise ValueError(f"Number of initial motifs ({len(initial_motifs)}) must match the number of DNA strings ({t}).")
    if any(len(motif) != k for motif in initial_motifs):
        raise ValueError(f"All initial motifs must have length k ({k}).")

    print(f"Initial Motifs: {initial_motifs}")

    # Step 2: Construct a profile matrix from these initial motifs.
    profile = profile_matrix_with_pseudocounts(initial_motifs)
    print("Profile Matrix (with pseudocounts):")
    for nucleotide, probabilities in profile.items():
        print(f"{nucleotide}: {probabilities}")

    # Step 3: For each DNA string, find the profile-most probable k-mer.
    new_motifs = [profile_most_probable_kmer(seq, k, profile) for seq in dna]
    print(f"Motifs after one iteration: {new_motifs}")

    return new_motifs
```

```python
import numpy as np
from scipy.stats import chi2_contingency

def compare_count_matrices_chi2(count_matrix_1, count_matrix_2):
    """
    Compares two count matrices (Position Frequency Matrices) using the Chi-squared test.
    It performs the Chi-squared test column by column and returns the average p-value.

    Args:
        count_matrix1 (numpy.ndarray): Count matrix 1 (shape: length x 4).
        count_matrix2 (numpy.ndarray): Count matrix 2 (shape: length x 4).
            Rows represent nucleotides (A, C, G, T), columns represent positions.

    Returns:
        float: The average p-value from the column-wise Chi-squared tests.
               A lower p-value indicates a greater difference between the columns.
    """
    if count_matrix_1.shape != count_matrix_2.shape:
        raise ValueError("Count matrices must have the same shape.")

    p_values = []
    for i in range(count_matrix_1.shape[0]):
        # Create a contingency table for the current column
        contingency_table = np.array([count_matrix_1[i, :], count_matrix_2[i, :]])
        # Perform the Chi-squared test
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        p_values.append(p)

    return np.mean(p_values)

def correct_for_multiple_comparisons(p_values, method='bonferroni'):
    """
    Corrects p-values for multiple comparisons using Bonferroni or Benjamini-Hochberg.

    Args:
        p_values (list of float): List of p-values.
        method (str): 'bonferroni' or 'fdr' (for Benjamini-Hochberg).
            Defaults to 'bonferroni'.

    Returns:
        list of float: List of corrected p-values.
    """
    if method == 'bonferroni':
        corrected_p_values = [min(1, p * len(p_values)) for p in p_values]
    elif method == 'fdr':
        from scipy.stats import rankdata
        ranked_p_values = rankdata(p_values, method='min')
        corrected_p_values = [
            min(1, p * len(p_values) / r) for p, r in zip(p_values, ranked_p_values)
        ]
    else:
        raise ValueError("Invalid method. Choose 'bonferroni' or 'fdr'.")
    return corrected_p_values
    
def compare_count_matrices_chi2_with_correction(count_matrix_1, count_matrix_2, correction_method='bonferroni'):
    """
    Compares two count matrices using column-wise Chi-squared tests, and corrects
    the resulting p-values for multiple comparisons.

    Args:
        count_matrix1 (numpy.ndarray): Count matrix 1 (shape: length x 4).
        count_matrix2 (numpy.ndarray): Count matrix 2 (shape: length x 4).
        correction_method (str): 'bonferroni' or 'fdr'.  Defaults to 'bonferroni'.

    Returns:
        tuple: (average_p_value, corrected_p_values, average_corrected_p_value)
            average_p_value (float): The average of the uncorrected p-values.
            corrected_p_values (list of float): The list of p-values, corrected for
                multiple comparisons.
            average_corrected_p_value (float): The average of the corrected p-values.
    """
    if count_matrix_1.shape != count_matrix_2.shape:
        raise ValueError("Count matrices must have the same shape.")

    p_values = []
    for i in range(count_matrix_1.shape[0]):
        contingency_table = np.array([count_matrix_1[i, :], count_matrix_2[i, :]])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        p_values.append(p)

    average_p_value = np.mean(p_values)
    corrected_p_values = correct_for_multiple_comparisons(p_values, method=correction_method)
    average_corrected_p_value = np.mean(corrected_p_values)

    return average_p_value, corrected_p_values, average_corrected_p_value
```

```python
import numpy as np
from scipy.stats import chi2_contingency

def compare_count_matrices_chi2(count_matrix_1, count_matrix_2):
    """
    Compares two count matrices (Position Frequency Matrices) using the Chi-squared test.
    It performs the Chi-squared test column by column and returns the average p-value.

    Args:
        count_matrix1 (numpy.ndarray): Count matrix 1 (shape: length x 4).
        count_matrix2 (numpy.ndarray): Count matrix 2 (shape: length x 4).
            Rows represent nucleotides (A, C, G, T), columns represent positions.

    Returns:
        float: The average p-value from the column-wise Chi-squared tests.
               A lower p-value indicates a greater difference between the columns.
               Returns 1.0 if a column comparison is invalid.
    """
    if count_matrix_1.shape != count_matrix_2.shape:
        raise ValueError("Count matrices must have the same shape.")

    p_values = []
    for i in range(count_matrix_1.shape[1]):  # Iterate over columns
        # Create a contingency table for the current column
        contingency_table = np.array([count_matrix_1[:, i], count_matrix_2[:, i]])
        # Perform the Chi-squared test
        try:
            chi2, p, dof, expected = chi2_contingency(contingency_table)
            p_values.append(p)
        except ValueError as e:
            print(f"Chi-squared test failed for column {i}: {e}")
            # Append a p-value of 1.0 for failed tests.  This will have minimal
            # impact on the *average* p-value, and indicates no difference for this comparison.
            p_values.append(1.0)

    return np.mean(p_values)



def correct_for_multiple_comparisons(p_values, method='bonferroni'):
    """
    Corrects p-values for multiple comparisons using Bonferroni or Benjamini-Hochberg.

    Args:
        p_values (list of float): List of p-values.
        method (str): 'bonferroni' or 'fdr' (for Benjamini-Hochberg).
            Defaults to 'bonferroni'.

    Returns:
        list of float: List of corrected p-values.
    """
    if method == 'bonferroni':
        corrected_p_values = [min(1, p * len(p_values)) for p in p_values]
    elif method == 'fdr':
        from scipy.stats import rankdata
        ranked_p_values = rankdata(p_values, method='min')
        corrected_p_values = [
            min(1, p * len(p_values) / r) for p, r in zip(p_values, ranked_p_values)
        ]
    else:
        raise ValueError("Invalid method. Choose 'bonferroni' or 'fdr'.")
    return corrected_p_values
    
def compare_count_matrices_chi2_with_correction(count_matrix_1, count_matrix_2, correction_method='bonferroni'):
    """
    Compares two count matrices using column-wise Chi-squared tests, and corrects
    the resulting p-values for multiple comparisons.

    Args:
        count_matrix1 (numpy.ndarray): Count matrix 1 (shape: length x 4).
        count_matrix2 (numpy.ndarray): Count matrix 2 (shape: length x 4).
        correction_method (str): 'bonferroni' or 'fdr'.  Defaults to 'bonferroni'.

    Returns:
        tuple: (average_p_value, corrected_p_values, average_corrected_p_value)
            average_p_value (float): The average of the uncorrected p-values.
            corrected_p_values (list of float): The list of p-values, corrected for
                multiple comparisons.
            average_corrected_p_value (float): The average of the corrected p-values.
    """
    if count_matrix_1.shape != count_matrix_2.shape:
        raise ValueError("Count matrices must have the same shape.")

    p_values = []
    for i in range(count_matrix_1.shape[1]):
        contingency_table = np.array([count_matrix_1[:, i], count_matrix_2[:, i]])
        try:
            chi2, p, dof, expected = chi2_contingency(contingency_table)
            p_values.append(p)
        except ValueError as e:
            print(f"Chi-squared test failed for column {i}: {e}")
            p_values.append(1.0)
            
    average_p_value = np.mean(p_values)
    corrected_p_values = correct_for_multiple_comparisons(p_values, method=correction_method)
    average_corrected_p_value = np.mean(corrected_p_values)

    return average_p_value, corrected_p_values, average_corrected_p_value
```
