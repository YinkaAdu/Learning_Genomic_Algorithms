```python
def composition(text, k):
   
    if not isinstance(text, str):
        raise TypeError("text must be a string")
    if not isinstance(k, int):
        raise TypeError("k must be an integer")
    if k <= 0:
        return ""  # Handle invalid k: no k-mers can be generated.
    if k > len(text):
        return ""  # Handle invalid k: k-mers cannot be longer than text.

    k_mers = []
    for i in range(len(text) - k + 1):
        k_mers.append(text[i:i+k])
    k_mers.sort()  # Sort the k-mers lexicographically.
    return " ".join(k_mers) # Join the k-mers with spaces.
```

```python
def reconstruct(kmer_strings, k):
    
    if not isinstance(kmer_strings, str):
        raise TypeError("kmer_string must be a string")
    if not isinstance(k, int):
        raise TypeError("k must be an integer")
    if k <= 0:
        return ""  # Handle invalid k
    if not kmer_strings:
        return ""

    kmer_strings = kmer_strings.split() # Split the space-separated string into a list

    # Check if all k-mers have the same length and equal to k
    for kmer in kmer_strings:
        if not isinstance(kmer, str) or len(kmer) != k:
            raise ValueError("All elements of kmer_strings must be strings of length k")

    # Sort the k-mers.  This is crucial for the algorithm to work correctly.
    kmer_strings.sort()

    # Initialize the reconstructed text with the first k-mer.
    text = kmer_strings[0]

    # Iterate through the remaining k-mers and extend the text.
    for i in range(1, len(kmer_strings)):
        # Overlap by k-1 characters.
        text += kmer_strings[i][-1]  # Add the last character of the next k-mer

    return text

kmer_strings = 'AAT  ATG  ATG  ATG  CAT  CCA  GAT  GCC  GGA  GGG  GTT  TAA  TGC  TGG  TGT'
k = 3
print(reconstruct(kmer_strings, k))

TAATGCCATGGATGTT
```
